```python
import random
import pandas as pd
pd.set_option('display.max_columns', None)
import time
from faker import Faker
fake = Faker('nl_NL')

# df = pd.read_csv('centres.csv')
# df.columns = ['centre']
# list = df["centre"].values.tolist()
# print(list)
```

```python
# list with options for the provider
referringCentres = ['Noordwest ziekenhuisgroep - Den Helder', 'Het Nederlands Kankerinstituut - Antoni van Leeuwenhoek ziekenhuis', 'Vumc ', 'Vumc - Dijklander ziekenhuis', 'AMC ', 'AMC - Flevoziekenhuis', 'Reinier de Graaf Groep', 'RCWest / Radiotherapiecentrum West', 'Haga Radiotherapie', 'Radiotherapiegroep - Deventer', 'Radiotherapiegroep - Arnhem', 'Radiotherapiegroep - Ede', 'Catharina Ziekenhuis', 'MST', 'Universitair Medisch Centrum Groningen / GPTC Groningen', 'Radiotherapeutisch Instituut Friesland', 'LUMC', 'MAASTRO   / MAASTRO Protonen', 'Radboudumc', 'ErasmusMC', 'ErasmusMC - Dordrecht', 'Instituut Verbeeten - Tilburg', 'Instituut Verbeeten - Breda', "Instituut Verbeeten - 's Hertogenbosch", 'Universitair Medisch Centrum Utrecht', 'PMC / Prinses Maxima Centrum', 'Zuidwest Radiotherapeutisch Instituut - Vlissingen', 'Zuidwest Radiotherapeutisch Instituut - Roosendaal', 'Radiotherapeutisch Centrum Zwolle', 'Holland PTC Delft', 'APTC Amsterdam']    
# first, import a similar Provider or use the default one
from faker.providers import BaseProvider
# create new provider class. Note that the class name _must_ be ``Provider``.
class Provider(BaseProvider):
    def GEN_REFER_CENTRE(self):
         return random.choice(referringCentres)
# then add new provider to faker instance
fake.add_provider(Provider)
# now you can use:
print(fake.GEN_REFER_CENTRE())

# gender
genders = ['male', 'female']
from faker.providers import BaseProvider
class Provider(BaseProvider):
    def GEN_GENDER(self):
         return random.choice(genders)
fake.add_provider(Provider)
print(fake.GEN_GENDER())

#age
from datetime import date, timedelta
def AGE(patient):
    fake.seed_instance(patient) #calling the same seed instance of faker creates the same results
    age = date.today() - fake.date_of_birth(tzinfo=None, minimum_age=35, maximum_age=101)
    return age.days/365
print(int(AGE(1)))

# educationLevel
educationLevels = ['Primary education (basisschool / lagere school)', "Preparatory vocational education (LBO/ LTS/ LEAO/ LHNO/ 'huishoudschool'/ VMBO-B/ VMBO-K)", 'General secondary education (MAVO/ (M)ULO/ MBO-kort/ VMBO-TL)', 'Secondary vocational education (MBO-lang/ MTS/ MEAO)', 'Higher general and preparatory scientific education (HAVO/ VWO/ HBS)', 'Higher professional education (HBO/ HTS/ HEAO)', 'Scientific education (WO/Universiteit)', 'Unknown', 'Other']    
from faker.providers import BaseProvider
class Provider(BaseProvider):
    def GEN_EDUCATION(self):
         return random.choice(educationLevels)
fake.add_provider(Provider)
print(fake.GEN_EDUCATION())

# relationshipStatus
relationshipStatus = ['Married', 'Never married', 'Domestic partner', 'Widowed']    
from faker.providers import BaseProvider
class Provider(BaseProvider):
    def GEN_RELATION(self):
         return random.choice(relationshipStatus)
fake.add_provider(Provider)
print(fake.GEN_RELATION())

# smokingStatus
def Smoking(patient):
    smokingStatus = ['No', 'Yes, in the past (ex-smoker)', 'Yes, current smoker']    
    packyears = 0
    stoppedMonths = 0
    random.seed(patient)
    status = random.choice(smokingStatus)
    if status != 'No':
        packyears = random.randint(1,50)
        stoppedMonths = random.randint(6,200)
    return [status, packyears, stoppedMonths] 
print(Smoking(1))

# alcoholStatus
def Alcohol(patient):
    alcoholStatus = ['No', 'Yes'] 
    setattr(Alcohol, 'drinkingDays', 0) #Everything in Python is an object. So functions have attributes like other objects
    Alcohol.glassesADay = 0 #or like this
    random.seed(patient)
    status = random.choice(alcoholStatus)
    if status != 'No':
        setattr(Alcohol, 'drinkingDays', random.randint(1,30))
        setattr(Alcohol, 'glassesADay',  random.randint(1,10))
    return status

print(Alcohol(1), Alcohol.drinkingDays, Alcohol.glassesADay)

# GEN_TUMOURSITE
tumourSite = ['Low grade glioma', 'Meningeoma', 'Base of skull', 'Craniospinal axis', 'Other tumour sites', 'Head and neck', 'Breast', 'Lung', 'Esophageal', 'Sarcoma', 'Lymphoma', 'Central nervous system', 'Craniospinal axis', 'Other tumour sites', 'Eye melanoma', 'Other tumour sites']    
from faker.providers import BaseProvider
class Provider(BaseProvider):
    def GEN_TUMOURSITE(self):
         return random.choice(tumourSite)
fake.add_provider(Provider)
print(fake.GEN_TUMOURSITE())

# GEN_RT_TREATMENT
def treatmentTechnique(patient):
    treatmentTechnique = ['Photons','Protons','Electrons','Intra Operative','Brachy']    
    random.seed(patient)
    return random.choice(treatmentTechnique)
print(treatmentTechnique(3))

# GEN_OTHERCANCER
def otherCancer(patient):
    cancer = ['Yes','No']    
    random.seed(patient)
    cancerBool = random.choice(cancer)
    otherCancer.cancerDate = 'Not applicable'
    otherCancer.cancerSite = 'Not applicable'
    otherCancer.previousRT = 'No'
    otherCancer.reIrradiation = 'No'
    if cancerBool == 'Yes':
        cancerTypes = ['Breast Cancer', 'Colon and Rectal Cancer', 'Endometrial Cancer', 'Kidney Cancer', 'Leukemia', 'Liver Cancer', 'Lung Cancer', 'Melanoma', 'Non-Hodgkin Lymphoma', 'Pancreatic Cancer', 'Prostate Cancer', 'Thyroid Cancer']
        otherCancer.cancerSite = random.choice(cancerTypes)
        otherCancer.cancerDate = random.randint(1960,2018)
        rTHistory = ['Yes', 'No', 'Unknown']
        otherCancer.previousRT = random.choice(rTHistory)
        if otherCancer.previousRT == 'Yes':
            otherCancer.irradiation = ['Yes', 'No', 'Unknown']
            otherCancer.reIrradiation = random.choice(otherCancer.irradiation)
    return (cancerBool)
print(otherCancer(3))

# GEN_OTHERCANCER
def planningComparison(patient):
    comparison = ['Yes','No'] 
    random.seed(patient)
    comparisonBool = random.choice(comparison)
    planningComparison.comparisonDate = 'Not applicable'
    planningComparison.comparisonVersion = 'Not applicable'
    planningComparison.comparisonOutcome = 'Not applicable'
    planningComparison.protons = 'Not applicable'
    planningComparison.noProtons = 'Not applicable'
    if comparisonBool == 'Yes':
        planningComparison.comparisonDate = fake.seed_instance(patient).date_between_dates(start_date=(2019,9,9), end_date=(2020,8,8) + timedelta(days=random.randint(2,20)))
        planningComparison.comparisonVersion = random.randint(1,3)
        if treatmentTechnique(patient) == 'Protons': 
            planningComparison.comparisonOutcome = 'Indication for proton therapy'
            planningComparison.protons = "yes"
        else: 
            planningComparison.comparisonOutcome = 'No indication for proton therapy'
            planningComparison.protons = 'No'
        noProtonsReason = ['Distance to proton centre', 'Benefit  proton therapy considered too limited', 'Other treatment team', 'Tumour progression', 'Other']
        planningComparison.noProtons = random.choice(noProtonsReason)
    return comparisonBool
print(planningComparison(8))

#do the same with a class and not a function
# class Alcohol():         
#         drinkingDays = 0
#         glassesADay = 0        
#         def __init__(self, patient): 
#             self.patient = patient            
#         def YesAlcohol(self):
#             alcoholStatus = ['No', 'Yes']    
#             random.seed(self.patient)
#             status = random.choice(alcoholStatus)
#             if status != 'No':
#                 self.drinkingDays = random.randint(1,30)
#                 self.glassesADay = random.randint(1,10)
#             return status  
```

```python
def create_rows(num=1):
    output = [{"GEN_IDNUMBER":55000101 + patient,
                "GEN_TREAT_CENTRE":"ZonPTC",
                "GEN_REFER_CENTRE":fake.GEN_REFER_CENTRE(),
                "GEN_DATE_REGISTR":fake.seed_instance(patient).date_between(start_date="-1y", end_date="today"),
                "GEN_GENDER":fake.GEN_GENDER(),
                "GEN_YEAR_BIRTH":fake.seed_instance(patient).date_of_birth(tzinfo=None, minimum_age=35, maximum_age=101),
                "GEN_AGE":int(AGE(patient)),
                "GEN_EDUCATION":fake.GEN_EDUCATION(),
                "GEN_RELATION":fake.GEN_RELATION(),
                "GEN_SMOKING":Smoking(patient)[0],
                "GEN_SM_PACKYEARS":Smoking(patient)[1],
                "GEN_SM_STOPMONTHS":Smoking(patient)[2],
                "GEN_ALCOHOL":Alcohol(patient),
                "GEN_ALCOHOL_DAYS":Alcohol.drinkingDays,
                "GEN_ALCOHOL_GLASS":Alcohol.glassesADay,
                "GEN_WEIGHT":random.randint(50,100),
                "GEN_HEIGHT":random.randint(165,200),
                "GEN_DATE_DIAGN":fake.seed_instance(patient).date_between(start_date="-1y", end_date="today") - timedelta(days=random.randint(2,60)),
                "GEN_TUMOURSITE":fake.GEN_TUMOURSITE(),
                "GEN_RT_TREATMENT":treatmentTechnique(patient),
                "GEN_OTHERCANCER":otherCancer(patient),
                "GEN_OTH_C_SITE":otherCancer.cancerSite,
                "GEN_OTH_C_DATE":otherCancer.cancerDate,
                "GEN_OTH_C_RT":otherCancer.previousRT,
                "GEN_OTH_C_RT_REIRR":otherCancer.reIrradiation,   
                "GEN_PLAN_COMPARISON":planningComparison(patient),
                "GEN_PLAN_COMP_DATE":planningComparison.comparisonDate,
                "GEN_PLAN_COMP_VERSION":planningComparison.comparisonVersion,
                "GEN_COMP_OUTCOME":planningComparison.comparisonOutcome,
                "GEN_DEC_PROTONS":planningComparison.protons,
                "GEN_REASON_PROTONS":planningComparison.noProtons,
              } for patient in range(num)]
    return output

#10 000 patients takes about 18sec
start_time = time.time()
df = pd.DataFrame(create_rows(1000))
elapsed_time = time.time() - start_time
print(elapsed_time)
```
